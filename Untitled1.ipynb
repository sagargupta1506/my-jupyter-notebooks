```python
import pandas as pd
data=pd.read_csv("diabetes.csv")
```

```python
data.head()
```

```python
X = data.drop('Outcome',axis=1)
X.head()
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

k=2
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X)

data['Cluster'] = clusters

print(data.sample(10))

```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=clusters, cmap='viridis')
plt.title('K-means Clusterinng(Without Normalization or PCA)')
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.show()
```

```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

k=2
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X)

data['Cluster'] = clusters

plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=clusters, cmap='viridis')
plt.title('K-means Clusterinng(With Normalization)')
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.show()
```

```python
scaler = StandardScaler()
X_normalized = scaler.fit_transform(X)

# Apply PCA to reduce the data to 2 components
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_normalized)

k=2
kmeans = KMeans(n_clusters=k, random_state=42)
clusters = kmeans.fit_predict(X)

data['Cluster'] = clusters

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=clusters, cmap='viridis')
plt.title('K-means Clusterinng(With Normalization and PCA)')
plt.xlabel('Pregnancies')
plt.ylabel('Glucose')
plt.show()
```
